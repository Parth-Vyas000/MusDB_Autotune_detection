```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
from google.colab import drive
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Conv2D, MaxPooling2D, Flatten, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from sklearn.model_selection import KFold
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score, confusion_matrix
import seaborn as sns
import librosa

# Mount Google Drive
drive.mount('/content/drive', force_remount=True)

# Define paths
train_dir = '/content/drive/MyDrive/Dla_Project/spectrogram_train'
test_dir = '/content/drive/MyDrive/Dla_Project/spectogram_test'

print("Checking directory paths...")
if not os.path.exists(train_dir):
    raise FileNotFoundError(f"Training directory not found: {train_dir}")
if not os.path.exists(test_dir):
    raise FileNotFoundError(f"Test directory not found: {test_dir}")

# Load and augment spectrogram
def load_spectrogram_image(image_path, target_size=(128, 431)):
    try:
        img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        if img is None:
            print(f"Failed to load image: {image_path}")
            return None
        img = cv2.resize(img, target_size[::-1])
        img = img.astype(np.float32) / 255.0
        # Augmentation: random noise, pitch shift simulation
        if np.random.random() > 0.5:
            noise = np.random.normal(0, 0.01, img.shape)
            img = np.clip(img + noise, 0, 1)
        return img
    except Exception as e:
        print(f"Error processing {image_path}: {e}")
        return None

# Load data
def load_data(data_dir):
    spectrograms = []
    labels = []
    folder_names = []
    folders = [f for f in os.listdir(data_dir) if os.path.isdir(os.path.join(data_dir, f))]
    folders.sort()
    num_folders = len(folders)
    labels = np.zeros(num_folders, dtype=int)
    labels[:num_folders//2] = 1
    for idx, folder_name in enumerate(folders):
        folder_path = os.path.join(data_dir, folder_name)
        vocals_path = os.path.join(folder_path, 'vocals.png')
        if os.path.exists(vocals_path):
            spec = load_spectrogram_image(vocals_path)
            if spec is not None:
                spectrograms.append(spec)
                folder_names.append(folder_name)
        else:
            print(f"Warning: vocals.png not found in {folder_path}")
    if len(spectrograms) != len(labels):
        print(f"Warning: Mismatch detected - {len(spectrograms)} spectrograms vs {len(labels)} labels")
        labels = labels[:len(spectrograms)]
    return np.array(spectrograms)[..., np.newaxis], np.array(labels), folder_names

train_spectrograms, train_labels, train_folder_names = load_data(train_dir)
print(f"Training Label Distribution: {np.bincount(train_labels)} (0: non-autotuned, 1: autotuned)")
print(f"Processed {len(train_spectrograms)} training samples.")

test_spectrograms, test_labels, test_folder_names = load_data(test_dir)
print(f"Test Label Distribution: {np.bincount(test_labels)} (0: non-autotuned, 1: autotuned)")
print(f"Processed {len(test_spectrograms)} test samples.")

# Cross-validation
kfold = KFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = []
all_train_history = []
all_metrics_history = []

for fold, (train_idx, val_idx) in enumerate(kfold.split(train_spectrograms)):
    print(f'Fold {fold + 1}')
    train_spec = train_spectrograms[train_idx]
    val_spec = train_spectrograms[val_idx]
    train_labels_split = train_labels[train_idx]
    val_labels_split = train_labels[val_idx]

    def create_dataset(specs, labels, batch_size=32, shuffle=True):
        dataset = tf.data.Dataset.from_tensor_slices((specs, labels))
        def process_data(spec, label):
            spec = tf.cast(spec, tf.float32)
            label = tf.cast(label, tf.float32)
            return spec, label
        dataset = dataset.map(process_data, num_parallel_calls=tf.data.AUTOTUNE)
        if shuffle:
            dataset = dataset.shuffle(buffer_size=1000)
        dataset = dataset.batch(batch_size)
        dataset = dataset.prefetch(tf.data.AUTOTUNE)
        return dataset

    train_dataset = create_dataset(train_spec, train_labels_split, shuffle=True)
    val_dataset = create_dataset(val_spec, val_labels_split, shuffle=False)

    # Model
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=(128, 431, 1)),
        BatchNormalization(),
        MaxPooling2D((2, 2)),
        Conv2D(64, (3, 3), activation='relu'),
        BatchNormalization(),
        MaxPooling2D((2, 2)),
        Conv2D(128, (3, 3), activation='relu'),
        BatchNormalization(),
        MaxPooling2D((2, 2)),
        Conv2D(256, (3, 3), activation='relu'),
        BatchNormalization(),
        MaxPooling2D((2, 2)),
        Flatten(),
        Dense(256, activation='relu'),
        Dropout(0.3),
        Dense(128, activation='relu'),
        Dropout(0.3),
        Dense(1, activation='sigmoid')
    ])

    model.compile(optimizer=Adam(learning_rate=1e-3), loss='binary_crossentropy', metrics=['accuracy'])

    class MetricsCallback(tf.keras.callbacks.Callback):
        def __init__(self, val_dataset):
            super(MetricsCallback, self).__init__()
            self.val_dataset = val_dataset
            self.precision = []
            self.recall = []
            self.f1 = []
            self.val_accuracy = []

        def on_epoch_end(self, epoch, logs=None):
            val_pred = []
            val_true = []
            for spec, labels in self.val_dataset:
                pred = (model.predict(spec, verbose=0) > 0.5).astype(int)
                val_pred.extend(pred.flatten())
                val_true.extend(labels.numpy().flatten())
            precision = precision_score(val_true, val_pred, zero_division=0)
            recall = recall_score(val_true, val_pred, zero_division=0)
            f1 = f1_score(val_true, val_pred, zero_division=0)
            self.precision.append(precision)
            self.recall.append(recall)
            self.f1.append(f1)
            self.val_accuracy.append(logs['val_accuracy'])
            print(f'Epoch {epoch+1}: Loss: {logs["loss"]:.4f}, Val Loss: {logs["val_loss"]:.4f}, '
                  f'Precision: {precision:.4f}, Recall: {recall:.4f}, F1: {f1:.4f}, Predictions: {np.bincount(val_pred)}')

    checkpoint = ModelCheckpoint(f'/content/drive/MyDrive/Dla_Project/best_model_fold{fold}.h5',
                                monitor='val_accuracy', save_best_only=True, mode='max', verbose=1)
    early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

    metrics_callback = MetricsCallback(val_dataset)
    history = model.fit(train_dataset, epochs=100, validation_data=val_dataset,
                       callbacks=[metrics_callback, checkpoint, early_stopping], verbose=1)
    all_train_history.append(history.history)
    all_metrics_history.append(metrics_callback.__dict__)

    # Evaluate on validation set
    val_pred = []
    val_true = []
    for spec, labels in val_dataset:
        pred = (model.predict(spec, verbose=0) > 0.5).astype(int)
        val_pred.extend(pred.flatten())
        val_true.extend(labels.numpy().flatten())
    accuracy = accuracy_score(val_true, val_pred)
    cv_scores.append(accuracy)
    print(f'Fold {fold + 1} Accuracy: {accuracy:.4f}\n')

print(f'Cross-validation Accuracy: {np.mean(cv_scores):.4f} (+/- {np.std(cv_scores)*2:.4f})')

# Fix for plotting: truncate histories to the shortest length
min_epochs = min(len(h['accuracy']) for h in all_train_history)
for i in range(len(all_train_history)):
    all_train_history[i]['accuracy'] = all_train_history[i]['accuracy'][:min_epochs]
    all_train_history[i]['val_accuracy'] = all_train_history[i]['val_accuracy'][:min_epochs]
    all_train_history[i]['loss'] = all_train_history[i]['loss'][:min_epochs]
    all_train_history[i]['val_loss'] = all_train_history[i]['val_loss'][:min_epochs]

# Average metrics across folds
avg_history = {}
for key in all_train_history[0].keys():
    avg_history[key] = np.mean([h[key] for h in all_train_history], axis=0)

# Average custom metrics
avg_precision = np.mean([h['precision'][:min_epochs] for h in all_metrics_history], axis=0)
avg_recall = np.mean([h['recall'][:min_epochs] for h in all_metrics_history], axis=0)
avg_f1 = np.mean([h['f1'][:min_epochs] for h in all_metrics_history], axis=0)

# Plot
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.plot(avg_history['accuracy'], label='Training Accuracy', color='green')
plt.plot(avg_history['val_accuracy'], label='Validation Accuracy', color='red', linestyle='--')
plt.title('Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(avg_precision, label='Validation Precision', color='green')
plt.title('Precision')
plt.xlabel('Epoch')
plt.ylabel('Precision')
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(avg_recall, label='Validation Recall', color='green')
plt.title('Recall')
plt.xlabel('Epoch')
plt.ylabel('Recall')
plt.legend()

plt.tight_layout()
plt.show()

# Final evaluation on test set using the best model from the last fold
best_model = tf.keras.models.load_model('/content/drive/MyDrive/Dla_Project/best_model_fold4.h5')  # Adjust fold index
test_dataset = create_dataset(test_spectrograms, test_labels, batch_size=32, shuffle=False)

predictions = []
true_labels = []
for spec, labels in test_dataset:
    pred = best_model.predict(spec, verbose=0)
    predictions.extend((pred > 0.5).astype(int).flatten())
    true_labels.extend(labels.numpy().flatten())

accuracy = accuracy_score(true_labels, predictions)
precision = precision_score(true_labels, predictions, zero_division=0)
recall = recall_score(true_labels, predictions, zero_division=0)
f1 = f1_score(true_labels, predictions, zero_division=0)
conf_matrix = confusion_matrix(true_labels, predictions)

print(f"\nTest Set Evaluation:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
print(f"Confusion Matrix:\n{conf_matrix}")

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Non-Autotuned', 'Autotuned'], yticklabels=['Non-Autotuned', 'Autotuned'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

folder_predictions = {}
for i, folder in enumerate(test_folder_names):
    if folder not in folder_predictions:
        folder_predictions[folder] = []
    folder_predictions[folder].append(predictions[i])

for folder, preds in folder_predictions.items():
    majority_vote = 1 if sum(preds) > len(preds) / 2 else 0
    print(f"Folder: {folder}, Predicted Label: {majority_vote} (1: Autotuned, 0: Non-Autotuned)")
```

```python
# Import necessary libraries
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
from google.colab import drive
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Conv2D, MaxPooling2D, Flatten
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score
import librosa.display

# Mount Google Drive (already mounted, but included for completeness)
drive.mount('/content/drive', force_remount=True)

# Define the path to your spectrogram_train folder
train_dir = '/content/drive/MyDrive/Dla_Project/spectrogram_train'

# Step 1: Load Spectrogram Images
# -------------------------------

# Function to load and preprocess spectrogram images
def load_spectrogram_image(image_path, target_size=(128, 431)):
    try:
        img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        if img is None:
            print(f"Failed to load image: {image_path}")
            return None
        # Resize to the target size (128x431 as in the paper)
        img = cv2.resize(img, target_size[::-1])  # OpenCV uses (width, height)
        # Normalize pixel values to [0, 1]
        img = img.astype(np.float32) / 255.0
        return img
    except Exception as e:
        print(f"Error processing {image_path}: {e}")
        return None

# Process all spectrogram images
spectrograms = []
labels = []
folder_names = []

# Get list of folders
folders = [f for f in os.listdir(train_dir) if os.path.isdir(os.path.join(train_dir, f))]
folders.sort()

# Temporary: Manually assign labels (50/50 split) since we can't apply Auto-Tune
num_folders = len(folders)
labels = np.zeros(num_folders, dtype=int)
labels[:num_folders//2] = 1  # First half as non-autotuned (0), second half as autotuned (1)

for idx, folder_name in enumerate(folders):
    folder_path = os.path.join(train_dir, folder_name)
    vocals_path = os.path.join(folder_path, 'vocals.png')  # Use spectrogram images
    if os.path.exists(vocals_path):
        spec = load_spectrogram_image(vocals_path)
        if spec is not None:
            spectrograms.append(spec)
            folder_names.append(folder_name)
    else:
        print(f"vocals.png not found in {folder_name}")

# Convert to numpy arrays
spectrograms = np.array(spectrograms)
labels = np.array(labels)

# Check if spectrograms list is empty
if len(spectrograms) == 0:
    raise ValueError("No spectrogram images were loaded. Please check the dataset path and file names.")

# Add channel dimension for CNN input (128x431x1)
spectrograms = spectrograms[..., np.newaxis]

# Print label distribution
print(f"Label Distribution: {np.bincount(labels)} (0: non-autotuned, 1: autotuned)")

# Save spectrograms and labels
np.save('/content/drive/MyDrive/Dla_Project/spectrograms.npy', spectrograms)
np.save('/content/drive/MyDrive/Dla_Project/labels.npy', labels)

# Plot an example spectrogram
example_spec = spectrograms[0].squeeze()
plt.figure(figsize=(10, 4))
plt.imshow(example_spec, aspect='auto', origin='lower', cmap='viridis')
plt.colorbar(format='%+2.0f')
plt.title('Spectrogram (Example)')
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.show()

# Print the number of processed samples
print(f"Processed {len(spectrograms)} samples.")

# Step 2: Train a CNN Model
# -------------------------

# Split the dataset into training and validation sets
train_idx, val_idx = train_test_split(
    np.arange(len(labels)), test_size=0.2, stratify=labels, random_state=42
)

train_spec = spectrograms[train_idx]
val_spec = spectrograms[val_idx]
train_labels = labels[train_idx]
val_labels = labels[val_idx]

print(f"Training set: {len(train_labels)} samples, Label distribution: {np.bincount(train_labels)}")
print(f"Validation set: {len(val_labels)} samples, Label distribution: {np.bincount(val_labels)}")

# Create tf.data.Dataset for training
def create_dataset(specs, labels, batch_size=32, shuffle=True):
    dataset = tf.data.Dataset.from_tensor_slices((specs, labels))

    def process_data(spec, label):
        spec = tf.cast(spec, tf.float32)
        label = tf.cast(label, tf.float32)
        return spec, label

    dataset = dataset.map(process_data, num_parallel_calls=tf.data.AUTOTUNE)
    if shuffle:
        dataset = dataset.shuffle(buffer_size=1000)
    dataset = dataset.batch(batch_size)
    dataset = dataset.prefetch(tf.data.AUTOTUNE)
    return dataset

# Create training and validation datasets
train_dataset = create_dataset(train_spec, train_labels, shuffle=True)
val_dataset = create_dataset(val_spec, val_labels, shuffle=False)

# Build a simple CNN model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(128, 431, 1)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(256, activation='relu'),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(
    optimizer=Adam(learning_rate=1e-5),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Custom callback to compute precision and recall
class MetricsCallback(tf.keras.callbacks.Callback):
    def __init__(self, val_dataset):
        super(MetricsCallback, self).__init__()
        self.val_dataset = val_dataset
        self.precision = []
        self.recall = []
        self.val_accuracy = []

    def on_epoch_end(self, epoch, logs=None):
        val_pred = []
        val_true = []
        for spec, labels in self.val_dataset:
            pred = (self.model.predict(spec, verbose=0) > 0.5).astype(int)
            val_pred.extend(pred.flatten())
            val_true.extend(labels.numpy().flatten())

        precision = precision_score(val_true, val_pred, zero_division=0)
        recall = recall_score(val_true, val_pred, zero_division=0)
        self.precision.append(precision)
        self.recall.append(recall)
        self.val_accuracy.append(logs['val_accuracy'])
        print(f'Epoch {epoch+1}: Loss: {logs["loss"]:.4f}, Val Loss: {logs["val_loss"]:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, Predictions: {np.bincount(val_pred)}')

# Define checkpoint callback to save the best model
checkpoint = ModelCheckpoint(
    '/content/drive/MyDrive/Dla_Project/best_model.h5',
    monitor='val_accuracy',
    save_best_only=True,
    mode='max',
    verbose=1
)

# Train the model
metrics_callback = MetricsCallback(val_dataset)
history = model.fit(
    train_dataset,
    epochs=30,
    validation_data=val_dataset,
    callbacks=[metrics_callback, checkpoint]
)

# Plot accuracy, precision, and recall
plt.figure(figsize=(15, 5))

# Accuracy
plt.subplot(1, 3, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy', color='green')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', color='red', linestyle='--')
plt.title('Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Precision
plt.subplot(1, 3, 2)
plt.plot(metrics_callback.precision, label='Validation Precision', color='green')
plt.title('Precision')
plt.xlabel('Epoch')
plt.ylabel('Precision')
plt.legend()

# Recall
plt.subplot(1, 3, 3)
plt.plot(metrics_callback.recall, label='Validation Recall', color='green')
plt.title('Recall')
plt.xlabel('Epoch')
plt.ylabel('Recall')
plt.legend()

plt.tight_layout()
plt.show()

# Save the final model
model.save('/content/drive/MyDrive/Dla_Project/final_model.h5')
```

```python
# Import necessary libraries
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
from google.colab import drive
import tensorflow as tf
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Dropout, Conv2D, MaxPooling2D, Flatten
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, accuracy_score, confusion_matrix
import seaborn as sns

# Mount Google Drive
drive.mount('/content/drive', force_remount=True)

# Define the paths to your spectrogram folders
train_dir = '/content/drive/MyDrive/Dla_Project/spectrogram_train'
test_dir = '/content/drive/MyDrive/Dla_Project/spectogram_test'

# Verify and correct paths
print("Checking directory paths...")
if not os.path.exists(train_dir):
    raise FileNotFoundError(f"Training directory not found: {train_dir}")
if not os.path.exists(test_dir):
    raise FileNotFoundError(f"Test directory not found: {test_dir}. Please check the folder name.")

# Step 1: Load and Preprocess Training Spectrogram Images
# ------------------------------------------------------

def load_spectrogram_image(image_path, target_size=(128, 431)):
    try:
        img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        if img is None:
            print(f"Failed to load image: {image_path}")
            return None
        img = cv2.resize(img, target_size[::-1])  # OpenCV uses (width, height)
        img = img.astype(np.float32) / 255.0
        return img
    except Exception as e:
        print(f"Error processing {image_path}: {e}")
        return None

# Load training spectrograms
train_spectrograms = []
train_labels = []
train_folder_names = []

train_folders = [f for f in os.listdir(train_dir) if os.path.isdir(os.path.join(train_dir, f))]
train_folders.sort()

num_folders = len(train_folders)
train_labels = np.zeros(num_folders, dtype=int)
train_labels[:num_folders//2] = 1  # Temporary 50/50 split

for idx, folder_name in enumerate(train_folders):
    folder_path = os.path.join(train_dir, folder_name)
    vocals_path = os.path.join(folder_path, 'vocals.png')
    if os.path.exists(vocals_path):
        spec = load_spectrogram_image(vocals_path)
        if spec is not None:
            train_spectrograms.append(spec)
            train_folder_names.append(folder_name)

train_spectrograms = np.array(train_spectrograms)
train_labels = np.array(train_labels)

if len(train_spectrograms) == 0:
    raise ValueError("No training spectrogram images were loaded.")

train_spectrograms = train_spectrograms[..., np.newaxis]

print(f"Training Label Distribution: {np.bincount(train_labels)} (0: non-autotuned, 1: autotuned)")
print(f"Processed {len(train_spectrograms)} training samples.")

# Step 2: Load and Preprocess Test Spectrogram Images
# --------------------------------------------------

test_spectrograms = []
test_folder_names = []

test_folders = [f for f in os.listdir(test_dir) if os.path.isdir(os.path.join(test_dir, f))]
test_folders.sort()

for folder_name in test_folders:
    folder_path = os.path.join(test_dir, folder_name)
    image_files = [f for f in os.listdir(folder_path) if f.endswith('.png')]
    for img_file in image_files:
        vocals_path = os.path.join(folder_path, img_file)
        spec = load_spectrogram_image(vocals_path)
        if spec is not None:
            test_spectrograms.append(spec)
            test_folder_names.append(folder_name)

# Dynamic label assignment based on the number of loaded images
num_test_images = len(test_spectrograms)
test_labels = np.zeros(num_test_images, dtype=int)
test_labels[:num_test_images//2] = 1  # 50/50 split across all loaded images

if len(test_spectrograms) == 0:
    raise ValueError("No test spectrogram images were loaded.")

test_spectrograms = np.array(test_spectrograms)[..., np.newaxis]

print(f"Test Label Distribution: {np.bincount(test_labels)} (0: non-autotuned, 1: autotuned)")
print(f"Processed {len(test_spectrograms)} test samples.")

# Step 3: Train the CNN Model
# ---------------------------

# Split training data for training and validation
train_idx, val_idx = train_test_split(
    np.arange(len(train_labels)), test_size=0.2, stratify=train_labels, random_state=42
)

train_spec = train_spectrograms[train_idx]
val_spec = train_spectrograms[val_idx]
train_labels_split = train_labels[train_idx]
val_labels_split = train_labels[val_idx]

def create_dataset(specs, labels, batch_size=32, shuffle=True):
    dataset = tf.data.Dataset.from_tensor_slices((specs, labels))
    def process_data(spec, label):
        spec = tf.cast(spec, tf.float32)
        label = tf.cast(label, tf.float32)
        return spec, label
    dataset = dataset.map(process_data, num_parallel_calls=tf.data.AUTOTUNE)
    if shuffle:
        dataset = dataset.shuffle(buffer_size=1000)
    dataset = dataset.batch(batch_size)
    dataset = dataset.prefetch(tf.data.AUTOTUNE)
    return dataset

train_dataset = create_dataset(train_spec, train_labels_split, shuffle=True)
val_dataset = create_dataset(val_spec, val_labels_split, shuffle=False)

# Build and train the model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(128, 431, 1)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(256, activation='relu'),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer=Adam(learning_rate=1e-5), loss='binary_crossentropy', metrics=['accuracy'])

class MetricsCallback(tf.keras.callbacks.Callback):
    def __init__(self, val_dataset):
        super(MetricsCallback, self).__init__()
        self.val_dataset = val_dataset
        self.precision = []
        self.recall = []
        self.val_accuracy = []

    def on_epoch_end(self, epoch, logs=None):
        val_pred = []
        val_true = []
        for spec, labels in self.val_dataset:
            pred = (model.predict(spec, verbose=0) > 0.5).astype(int)
            val_pred.extend(pred.flatten())
            val_true.extend(labels.numpy().flatten())

        precision = precision_score(val_true, val_pred, zero_division=0)
        recall = recall_score(val_true, val_pred, zero_division=0)
        self.precision.append(precision)
        self.recall.append(recall)
        self.val_accuracy.append(logs['val_accuracy'])
        print(f'Epoch {epoch+1}: Loss: {logs["loss"]:.4f}, Val Loss: {logs["val_loss"]:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, Predictions: {np.bincount(val_pred)}')

checkpoint = ModelCheckpoint(
    '/content/drive/MyDrive/Dla_Project/best_model.h5',
    monitor='val_accuracy',
    save_best_only=True,
    mode='max',
    verbose=1
)

metrics_callback = MetricsCallback(val_dataset)
history = model.fit(
    train_dataset,
    epochs=30,
    validation_data=val_dataset,
    callbacks=[metrics_callback, checkpoint]
)

# Plot training metrics
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy', color='green')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', color='red', linestyle='--')
plt.title('Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(metrics_callback.precision, label='Validation Precision', color='green')
plt.title('Precision')
plt.xlabel('Epoch')
plt.ylabel('Precision')
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(metrics_callback.recall, label='Validation Recall', color='green')
plt.title('Recall')
plt.xlabel('Epoch')
plt.ylabel('Recall')
plt.legend()

plt.tight_layout()
plt.show()

# Save the final model
model.save('/content/drive/MyDrive/Dla_Project/final_model.h5')

# Step 4: Evaluate on Test Dataset
# --------------------------------

# Load the best or final model
model = load_model('/content/drive/MyDrive/Dla_Project/final_model.h5')

# Create test dataset
test_dataset = create_dataset(test_spectrograms, test_labels, batch_size=32, shuffle=False)

# Make predictions
predictions = []
true_labels = []
for spec, labels in test_dataset:
    pred = model.predict(spec, verbose=0)
    predictions.extend((pred > 0.5).astype(int).flatten())
    true_labels.extend(labels.numpy().flatten())

# Compute evaluation metrics
accuracy = accuracy_score(true_labels, predictions)
precision = precision_score(true_labels, predictions, zero_division=0)
recall = recall_score(true_labels, predictions, zero_division=0)
conf_matrix = confusion_matrix(true_labels, predictions)

print(f"\nTest Set Evaluation:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"Confusion Matrix:\n{conf_matrix}")

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Non-Autotuned', 'Autotuned'], yticklabels=['Non-Autotuned', 'Autotuned'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# Optional: Print per-folder predictions (aggregate predictions by folder)
folder_predictions = {}
for i, folder in enumerate(test_folder_names):
    if folder not in folder_predictions:
        folder_predictions[folder] = []
    folder_predictions[folder].append(predictions[i])

for folder, preds in folder_predictions.items():
    majority_vote = 1 if sum(preds) > len(preds) / 2 else 0  # Majority vote
    print(f"Folder: {folder}, Predicted Label: {majority_vote} (1: Autotuned, 0: Non-Autotuned)")
```

```python
import subprocess  # To convert MP4 to WAV

# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Define dataset paths
TRAIN_PATH = "/content/drive/MyDrive/train"
TEST_PATH = "/content/drive/MyDrive/test"
OUTPUT_CSV_PATH = "/content/drive/MyDrive/musdb18_csv"
TEMP_WAV_PATH = "/content/temp_audio.wav"  # Temporary WAV file for processing

# Ensure output directory exists
os.makedirs(OUTPUT_CSV_PATH, exist_ok=True)

# Function to convert MP4 to WAV using FFmpeg
def convert_mp4_to_wav(mp4_path, wav_path):
    try:
        command = f"ffmpeg -i \"{mp4_path}\" -ac 1 -ar 22050 \"{wav_path}\" -y"
        subprocess.run(command, shell=True, check=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        return True
    except subprocess.CalledProcessError as e:
        print(f"❌ Error converting {mp4_path}: {e}")
        return False

# Function to extract pitch and save CSV
def extract_pitch_and_save(audio_path, csv_output_path):
    try:
        # Convert MP4 to WAV
        if audio_path.endswith(".mp4"):
            success = convert_mp4_to_wav(audio_path, TEMP_WAV_PATH)
            if not success:
                return  # Skip this file if conversion failed
            audio_path = TEMP_WAV_PATH  # Use converted WAV file

        # Load the audio
        y, sr = librosa.load(audio_path, sr=22050)

        # Extract pitch using PYIN
        pitches, voiced_flags, _ = librosa.pyin(
            y, fmin=librosa.note_to_hz('C2'), fmax=librosa.note_to_hz('C7')
        )

        # Check if pitch extraction was successful
        if pitches is None or np.all(np.isnan(pitches)):
            print(f"⚠️ No pitch detected in {audio_path}")
            return

        # Create DataFrame
        times = librosa.times_like(pitches, sr=sr)
        pitch_data = pd.DataFrame({"Time (s)": times, "Pitch (Hz)": pitches})

        # Remove NaN values
        pitch_data = pitch_data.dropna()

        # Save to CSV
        pitch_data.to_csv(csv_output_path, index=False)
        return True  # Success

    except Exception as e:
        print(f"❌ Error processing {audio_path}: {str(e)}")
        return False  # Failure

# Get the list of MP4 files from Train & Test directories
file_list = []
for folder in [TRAIN_PATH, TEST_PATH]:
    for filename in os.listdir(folder):
        if filename.endswith(".mp4"):
            file_list.append((folder, filename))

# Process all MP4 files with tqdm progress bar
with tqdm(total=len(file_list), desc="Processing Songs", unit="file") as pbar:
    for folder, filename in file_list:
        audio_path = os.path.join(folder, filename)
        csv_output_path = os.path.join(OUTPUT_CSV_PATH, filename.replace(".mp4", ".csv"))
        success = extract_pitch_and_save(audio_path, csv_output_path)

        if success:
            pbar.set_postfix(status="✅ Done")
        else:
            pbar.set_postfix(status="❌ Failed")

        pbar.update(1)  # Update progress bar

print("🎉 All songs processed & saved as CSV!")
```

```python
import os
import torch
import pandas as pd
from tqdm import tqdm
from torch.utils.data import Dataset, DataLoader
from torch.nn.utils.rnn import pad_sequence

class AutoTuneDataset(Dataset):
    def __init__(self, folder_path):
        self.files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith('.csv')]
        self.data = []

        print(f"Loading data from {folder_path}...")
        for file in tqdm(self.files, desc="Processing Files", unit="file"):
            df = pd.read_csv(file)
            features = df[['Pitch (Hz)', 'Auto-Tuned Pitch (Hz)']].values
            labels = df['Label'].values
            self.data.append((torch.tensor(features, dtype=torch.float32), torch.tensor(labels, dtype=torch.long)))

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return self.data[idx]

def collate_fn(batch):
    inputs, labels = zip(*batch)

    # Get sequence lengths
    lengths = torch.tensor([len(seq) for seq in inputs], dtype=torch.long)

    # Sort by sequence length (required for LSTM)
    sorted_indices = torch.argsort(lengths, descending=True)
    inputs = [inputs[i] for i in sorted_indices]
    labels = [labels[i] for i in sorted_indices]
    lengths = lengths[sorted_indices]

    # Convert to padded tensors
    inputs = pad_sequence(inputs, batch_first=True, padding_value=0.0)
    labels = pad_sequence(labels, batch_first=True, padding_value=-1)  # -1 for ignored values in loss

    return inputs, lengths, labels

# Define dataset and dataloader
train_dataset = AutoTuneDataset('/content/drive/MyDrive/musdb18_autotuned_csv_lbled/train')
test_dataset = AutoTuneDataset('/content/drive/MyDrive/musdb18_autotuned_csv_lbled/test')

train_loader = DataLoader(train_dataset, batch_size=32, collate_fn=collate_fn, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, collate_fn=collate_fn, shuffle=False)

# Verify dataset loading with tqdm
print("Verifying dataset loading...")
for batch in tqdm(train_loader, desc="Loading Batches", unit="batch"):
    inputs, lengths, labels = batch
    print(f"Batch Input Shape: {inputs.shape}")
    print(f"Sequence Lengths: {lengths}")
    print(f"Labels Shape: {labels.shape}")
    break
```

```python
# Define paths
input_root = '/content/drive/MyDrive/musdb18_autotuned_csv_lbled'  # Change this
output_root = '/content/drive/MyDrive/musdb18_autotuned_csv_lbled/processed'

input_train = os.path.join(input_root, 'train')
input_test = os.path.join(input_root, 'test')
output_train = os.path.join(output_root, 'train')
output_test = os.path.join(output_root, 'test')

# Make output folders if not exist
os.makedirs(output_train, exist_ok=True)
os.makedirs(output_test, exist_ok=True)

# Majority Voting Function (Sliding Window)
def add_majority_label(df, window_size=5):
    labels = df['Label'].tolist()
    majority_labels = []

    for i in range(len(labels)):
        start = max(0, i - window_size + 1)
        window = labels[start:i+1]
        majority = Counter(window).most_common(1)[0][0]
        majority_labels.append(majority)

    df['Window_Label'] = majority_labels
    return df

# Process files in a folder
def process_folder(input_folder, output_folder):
    for file in os.listdir(input_folder):
        if file.endswith('.csv'):
            file_path = os.path.join(input_folder, file)
            df = pd.read_csv(file_path)

            df = add_majority_label(df, window_size=5)

            output_path = os.path.join(output_folder, file)
            df.to_csv(output_path, index=False)
            print(f"Processed: {file}")

# Process train and test folders
process_folder(input_train, output_train)
process_folder(input_test, output_test)
```
